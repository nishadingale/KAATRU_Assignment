### Significant_varible_finder function

```python
# Below is a function that takes in a dataframe,dependant variable and correlation - 
# value to classify an independant variable as significant or not and return a dataframe of correlation values and names of the variables
def significant_varible_finder(df,dependant_variable,correlation_value = 0.4):    
    high_corr = []
    corr_value = {}
    for var in df.columns:
        cor = df["cnt"].corr(df[var])
        if cor > correlation_value:
            high_corr.append([var])
            corr_value[var]= cor

    corr_values = pd.DataFrame.from_dict(corr_value,orient = "index",columns = [f"Correlation with {dependant_variable}"])

    return corr_values
```

```python
data = significant_varible_finder(df,'cnt')
```
